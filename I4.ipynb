method 1 : Kaggle Datase

```python
import csv


def read_csv_local(filepath: str = "bestsellers_with_categories_2022_03_27.csv"):

    """
    PROS:
        - No internet connection needed at runtime; data is already on disk.
        - Full control over the dataset — you can inspect, clean, or version
          it before your code ever touches it.
        - Very fast reads since there is no network latency.

    CONS:
        - Data is static — if the source updates, you must re-download manually.
        - Requires a manual download step before the script can run, which adds
          friction and makes full automation harder.
        - Large files consume local disk space.
        - The CSV format itself has no strict schema, so malformed rows,
          inconsistent quoting, or encoding issues are common.
    """
    print("=" * 70)
    print("METHOD 1: CSV — Local File Read (Kaggle Amazon Bestsellers)")
    print("=" * 70)

    # Open and parse the CSV using DictReader so each row is a dictionary
    with open(filepath, newline="", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        rows = list(reader)

    # Print dataset overview
    print(f"   Total records loaded: {len(rows)}")
    print(f"   Columns: {list(rows[0].keys())}\n")

    # Print a sample of the data
    print("   Sample — First 5 Books:")
    print("   " + "-" * 66)

    for i, row in enumerate(rows[:5]):
        title = row["Name"][:50] + "..." if len(row["Name"]) > 50 else row["Name"]
        print(f"   {i+1}. {title}")
        print(f"      Author: {row['Author']}  |  Rating: {row['User Rating']}  "
              f"|  Price: ${row['Price']}  |  Genre: {row['Genre']}")

    print()
    return rows


if __name__ == "__main__":
    print()
    data = read_csv_local()
    print(f"Successfully loaded {len(data)} records from local CSV file.")
```

method 2: HTML — Web Scraping on Wikipedia

```python
# --- Install dependencies ---
import subprocess
subprocess.check_call(["pip", "install", "-q", "beautifulsoup4"])

import requests
from bs4 import BeautifulSoup


def scrape_coffee_production(
    url="https://en.wikipedia.org/wiki/List_of_countries_by_coffee_production"
):
    """
    Scrape the Wikipedia "List of countries by coffee production" page and
    extract the HTML table of countries and their production volumes using
    BeautifulSoup.

    ACCESS TECHNOLOGY: HTTP GET to download raw HTML, then parse it with
    the BeautifulSoup library to extract specific elements from the DOM.

    INFORMATION STRUCTURE: HTML (HyperText Markup Language) — a tree-based
    document format that mixes content with presentation markup, requiring
    a parser to extract the data you actually want.

    PROS:
        - Can extract data from virtually any public website, even those
          that do not offer an API.
        - BeautifulSoup provides powerful CSS-selector and tag-based search,
          making it easy to target specific page elements.
        - Relatively low barrier to entry — just requests + BeautifulSoup.

    CONS:
        - Fragile: if the website changes its HTML structure or class names,
          the scraper breaks with no warning.
        - May violate a site's Terms of Service or robots.txt; always check
          before scraping.
        - Raw HTML is noisy — it mixes content with layout, ads, navigation,
          and scripts, so significant parsing logic is needed.
    """

    print("=" * 70)
    print("METHOD 2: HTML — Web Scraping (Wikipedia: Coffee Production)")
    print("=" * 70)

    response = requests.get(
        url, timeout=10, headers={"User-Agent": "StudentProject/1.0"}
    )
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "html.parser")

    page_title = soup.find("h1").get_text()
    print(f"   Page title: {page_title}")
    print(f"   HTML size:  {len(response.text):,} characters\n")

    table = soup.find("table", class_="wikitable")
    countries = []

    if table:
        rows = table.find_all("tr")
        for row in rows[1:]:
            cells = row.find_all("td")
            if len(cells) >= 2:
                country = cells[0].get_text().strip()
                production = cells[1].get_text().strip()
                countries.append({"country": country, "production_tonnes": production})

    print(f"   Countries found: {len(countries)}")
    print("   Top 10 Coffee Producing Countries (2023):")
    print("   " + "-" * 66)
    print(f"   {'Rank':<6} {'Country':<30} {'Production (tonnes)':>20}")
    print("   " + "-" * 66)

    for i, entry in enumerate(countries[:10], 1):
        print(f"   {i:<6} {entry['country']:<30} {entry['production_tonnes']:>20}")

    print()
    return countries

scrape_coffee_production()
```

method 3: API

```python
import json
import requests


def fetch_weather_api(
    latitude=47.6062,
    longitude=-122.3321,
    city_name="Seattle"
):
    """
    Fetch current weather and a 7-day daily forecast from the Open-Meteo API.

    ACCESS TECHNOLOGY: HTTP GET request to a RESTful API endpoint that
    returns structured JSON data. No API key or authentication required.

    INFORMATION STRUCTURE: JSON (JavaScript Object Notation) — a lightweight,
    human-readable format that maps directly to Python dicts and lists,
    making it the most common format for modern web APIs.

    PROS:
        - Always returns the latest data — no manual refresh needed.
        - JSON is lightweight, human-readable, and natively maps to Python
          dicts/lists, making parsing trivial with response.json().
        - REST APIs are the industry standard; thousands of public and
          private services expose data this way.
        - Stateless requests are easy to cache, retry, and parallelize.
        - Many APIs support query parameters for filtering, pagination, and
          sorting, so you can request only the data you need.

    CONS:
        - Requires an active internet connection at runtime.
        - Subject to rate limits, downtime, or breaking API changes that are
          outside your control.
        - Many real-world APIs require authentication (API keys, OAuth), which
          adds setup complexity and key-management overhead.
        - Network latency can be significant for large payloads or slow
          servers.
        - You are dependent on the provider's data schema — if they rename a
          field or deprecate an endpoint, your code breaks.
    """
    print("=" * 70)
    print(f"METHOD 3: JSON — REST API (Open-Meteo Weather for {city_name})")
    print("=" * 70)

    # Build the API request URL with query parameters
    url = "https://api.open-meteo.com/v1/forecast"
    params = {
        "latitude": latitude,
        "longitude": longitude,
        "current": "temperature_2m,relative_humidity_2m,wind_speed_10m,weather_code",
        "daily": "temperature_2m_max,temperature_2m_min,precipitation_sum",
        "temperature_unit": "fahrenheit",
        "wind_speed_unit": "mph",
        "precipitation_unit": "inch",
        "timezone": "America/Los_Angeles",
        "forecast_days": 7
    }

    response = requests.get(url, params=params, timeout=10)
    response.raise_for_status()
    data = response.json()

    # Print raw JSON structure info
    print(f"   HTTP Status: {response.status_code}")
    print(f"   JSON keys:   {list(data.keys())}")
    print(f"   Timezone:    {data.get('timezone', 'N/A')}\n")

    # Print current conditions
    current = data.get("current", {})
    print(f"   Current Weather in {city_name}:")
    print("   " + "-" * 66)
    print(f"   Temperature:     {current.get('temperature_2m', 'N/A')}°F")
    print(f"   Humidity:        {current.get('relative_humidity_2m', 'N/A')}%")
    print(f"   Wind Speed:      {current.get('wind_speed_10m', 'N/A')} mph")
    print(f"   Weather Code:    {current.get('weather_code', 'N/A')}")
    print()

    # Print 7-day forecast
    daily = data.get("daily", {})
    dates = daily.get("time", [])
    highs = daily.get("temperature_2m_max", [])
    lows = daily.get("temperature_2m_min", [])
    precip = daily.get("precipitation_sum", [])

    print(f"   7-Day Forecast for {city_name}:")
    print("   " + "-" * 66)
    print(f"   {'Date':<14} {'High (°F)':>10} {'Low (°F)':>10} {'Precip (in)':>12}")
    print("   " + "-" * 66)

    for i in range(len(dates)):
        print(f"   {dates[i]:<14} {highs[i]:>10.1f} {lows[i]:>10.1f} {precip[i]:>12.2f}")

    print()
    return data


# *** Run the function ***
fetch_weather_api()
```
